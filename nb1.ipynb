```python
word = "Mallikarjuna"
vowels = "aeiouAEIOU"
vowel_count = 0
for char in word:
    if char in vowels:
        vowel_count += 1
print(f"The word '{word}' has {vowel_count} vowels.")
```

```python
vowel = ['a', 'e', 'i', 'o', 'u']
word = "mallikarjuna"
count = 0
for i in word:
    if i in vowel:
        count += 1
print(count)
print(f"the word '{word}' has {count} vowels.")
```
