counting consonants in a given word


```python
vowel = ['a', 'e', 'i', 'o', 'u']
word = "programming"
count = 0
for charecter in word:
  if charecter not in vowel:
    count += 1
print(count)
```

counting the number of occurances of a charectoer in string

```python
word = "programming"
charecter = "g"
count = 0
for i in word:
  if i == charecter:
    count +=1
print(count)
```

writing fibonacci series

```python
fib = [0,1]
for i in range(2, 10):
  next_fib = fib[i-1] + fib[i-2]
  fib.append(next_fib)
print(fib)
```
